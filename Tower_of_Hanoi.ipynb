## Importing the libraries

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import time
```

## Setting the configurationsmm

```python
NUM_DISKS = 4
SLEEP_TIME = 0.5  #Delay between moves (seconds)

towers = {
    'A': list(reversed(range(1, NUM_DISKS + 1))),
    'B': [],
    'C': []
}
```

##Functions

```python
def draw_towers(step):
    plt.clf()
    rod_positions = {'A': 0, 'B': 1, 'C': 2}
    colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan', 'yellow']

    # Draw rods
    for i in range(3):
        plt.plot([i, i], [0, NUM_DISKS + 1], color='black', linewidth=2)

    # Draw disks
    for rod in ['A', 'B', 'C']:
        stack = towers[rod]
        for level, disk in enumerate(stack):
            x_center = rod_positions[rod]
            width = disk * 0.15
            height = 0.5
            y = level * height
            rect = patches.Rectangle((x_center - width / 2, y), width, height, color=colors[disk % len(colors)])
            plt.gca().add_patch(rect)

    plt.xlim(-1, 3)
    plt.ylim(0, NUM_DISKS + 1)
    plt.xticks([0, 1, 2], ['A', 'B', 'C'])
    plt.title(f"Step {step}")
    plt.pause(SLEEP_TIME)
```

```python
def move(n, source, target, auxiliary, step_tracker):
    if n == 1:
        disk = towers[source].pop()
        towers[target].append(disk)
        step_tracker[0] += 1
        draw_towers(step_tracker[0])
    else:
        move(n - 1, source, auxiliary, target, step_tracker)
        move(1, source, target, auxiliary, step_tracker)
        move(n - 1, auxiliary, target, source, step_tracker)
```

```python
def solve_tower_of_hanoi():
    step_tracker = [0]
    plt.ion()
    draw_towers(0)
    move(NUM_DISKS, 'A', 'C', 'B', step_tracker)
    plt.ioff()
    plt.show()
    print(f"Total moves: {step_tracker[0]} (Expected: {2**NUM_DISKS - 1})")
```

## Calling functions

```python
solve_tower_of_hanoi()
```
